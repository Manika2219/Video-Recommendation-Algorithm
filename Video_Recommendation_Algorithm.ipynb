
# Function to fetch data from an API with pagination

```python
def fetch_data(api_url):
    data = []
    page = 1

    while True:
        url = f"{api_url}&page={page}"
        response = requests.get(url, headers=headers)
        print(f"Fetching page {page}: Status code {response.status_code}")
         # Check for HTTP errors
        if response.status_code != 200:
            print(f"Error: {response.status_code}")
            break

        try:
            page_data = response.json()
            print(json.dumps(page_data, indent=2))  # Inspect the response structure

            # Check for valid data keys in the response
            if 'data' in page_data and page_data['data']:
                data.extend(page_data['data'])
            elif 'posts' in page_data and page_data['posts']:
                data.extend(page_data['posts'])
            elif 'items' in page_data and page_data['items']:
                data.extend(page_data['items'])
            else:
                print("No more data or unrecognized response format.")
                break

            page += 1
            time.sleep(1)  # Sleep to respect rate limits if any

        except json.JSONDecodeError:
            print("Error decoding the JSON response.")
            break

    return pd.DataFrame(data)

# API URL to fetch posts
posts_api = "https://api.socialverseapp.com/posts/summary/get?page_size=1000"

# Fetch the data
posts_df = fetch_data(posts_api)

# Display the first few rows
if not posts_df.empty:
    print(posts_df.head())
else:
    print("No data was fetched.")
```
